## Program 1: Swap Two Numbers

**Problem**: Read two integers `a` and `b` and swap them.

**Input**:
- Two integers `a` and `b` (space-separated)

**Output**:
- Print the swapped values as: `a b`

**Sample Input**:
```
10 25
```
**Sample Output**:
```
25 10
```


````python
a,b=map(int,input("Enter the two numbers :").split(" "))
print(f"Sample input: --> {a} {b}")
# --------------------------------
# Swap using Python unpacking
# --------------------------------
a,b=b,a
print(f"Swap Output using unpacking {a} b}")
# Revert values for next method
b,a=a,b
# ---------------------------------
# Swap using a temporary variable
# ---------------------------------
c=a+b-a
b=a
a=c
print("Swap Output (using temp variable):", a, b)
# Revert values for next method
a,b=b,a
# ---------------------------------
# Swap using arithmetic operations
#----------------------------------
a=a+b #(a+b==>10+25=35==>a=35)
b=a-b # (a-b==>35-25=10==> b=10)
a=a-b #(a-b==>35-10=25==>a=25)

print("Swap Output (arithmetic method):", a, b)
````

## Program 4: Simple Interest

**Problem**: Read `P`, `R`, `T` and compute Simple Interest.

**Formula**: SI = (P * R * T) / 100

**Input**:
- Three numbers `P R T` (space-separated)

**Output**:
- Print `SI`

**Sample Input**:
```
1000 5 2
```
**Sample Output**:
```
100.0
```


````python
# Read principal, interest rate, and time from user as floats
principal,interest,time=map(float,input("Enter the principal ,intrest ,and time in order :").split(" "))
print(f"principal: {principal}\nintrest : {interest}%\ntime : {time} years")
# Simple Interest formula:
# SI = (P × R × T) / 100
# P = principal amount
# R = annual interest rate
# T = time in years
simple_interest= (principal * interest * time)/100
print(f"calculated simple interest for the given details : {simple_interest}")
````

## Program 7: Largest of Three Numbers

**Problem**: Read three integers and print the largest.

**Sample Output**:

## Program 8: Leap Year Check

**Problem**: Read a year and print `Leap Year` or `Not Leap Year`.

**Rule**:
- Leap if divisible by 400 OR divisible by 4 but not by 100.

**Input**:
- One integer `year`

**Output**:
- `Leap Year` or `Not Leap Year`

**Sample Input**:
```
2024
```
**Sample Output**:
```
Leap Year
```


````python
# Read the year from the user as an integer
year = int(input("Enter the year : "))
print(f"Entered years is {year}")
# Check if the year is a leap year using the standard rules:
# if the year is divisible by 4 and not divisible by 100 -->leap year (regular)
# if the year is divisible by 100 then it should be divisble by 400 also to be leap year (century)
# for century years if its divisible by 400 then its leap year
if (year % 4 ==0 and  year % 100 !=0  or year % 400==0 ):
  print(f"Entered year {year} is 'Leap Year'")
else:
  print(f"Entered year {year} is 'Not a Leap Year'")
````

## Program 9: Grade from Marks

**Problem**: Read marks (0–100) and print grade.

**Grading**:
- 90–100: A
- 80–89: B
- 70–79: C
- 60–69: D
- <60: F

**Input**:
- One integer `marks`

**Output**:
- Print the grade letter

**Sample Input**:
```
83
```
**Sample Output**:
```
B
```


````python
# Read the marks from user as integar
mark = int(input("Enter the mark to check for GRADE :"))
print(f"Entered mark is {mark}")
# Determine the grade based on the score
# Grade A → 90 and above
# Grade B → 80 to 89
# Grade C → 70 to 79
# Grade D → 60 to 69
# Grade F → below 60
if mark >=90:
  print("GRADE-A")
elif mark >=80:
  print("GRADE-B")
elif mark >=70:
  print("GRADE-C")
elif mark >=60:
  print("GRADE-D")
else:
  print("GRADE-F")
````
