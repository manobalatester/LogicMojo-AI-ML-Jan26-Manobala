**Input**:
- One integer `N` (assume `N >= 0`)

**Output**:
- Print factorial

**Sample Input**:
```
5
```

```

```


````python
# Read number to find factorial of it from user
number = int(input("Enter a number to find its factorial: "))
print(f"Entered number is  {number}")
# factorial formula
# n!=n(n-1)(n-2)(n-3)
# Loop from 'number' down to 1
# Example: for number = 5!→ 5, 4, 3, 2, 1
factorial=1
for i in range(number,0,-1):
  # Multiply factorial by the current number 'i'
  # i=5 ==>1*5=>factorial =5
  # i=4 ==>5*4=>factorial =20
  # i=3 ==>20*3=>factorial =60
  # i=2 ==>60*2=>factorial =120
  # i=1 ==>120*1=>factorial =120
    factorial=factorial*i
print(f"Factorial of {number}!--> {factorial}")


````

## Program 14: Prime Check

**Problem**: Read an integer `n` and print `Prime` or `Not Prime`.

**Input**:
- One integer `n`

**Output**:
- `Prime` or `Not Prime`

**Sample Input**:
```
29
```
**Sample Output**:
```
Prime
```


````python
# Read the integar from user where it neeeds to be cheked for prime or not
number = int(input("Enter the number to be checked for prime or not :"))
print(f"Entered number is : {number}")
# count howmany time number is divisible
count=0
# we start the range from 2 to till number-1 since 1 and same number is always divisible by it
import math
for i in range(2,int(math.sqrt(number))+1):
  if number % i == 0: # if divisible, it's not prime
    count+=1
# If count is 1 or more then given number is not PRIME
if count >= 1 :
  print(f"Entered number is : {number} is not Prime")
else:
  print(f"Entered number is : {number} is Prime")
````

## Program 26: Reverse a List (without reverse())

**Problem**: Read `N` and a list. Print the list in reverse order without using `.reverse()`.

**Input**:
- Integer `N`
- Next line: `N` integers

**Output**:
- Reversed list elements separated by space

**Sample Input**:
```
5
10 20 30 40 50
```
**Sample Output**:
```
50 40 30 20 10
```


````python
# read the integar from user
number=int(input("Enter the total length of list :"))
number_list=list(map(int,(input("Enter the list of numbers in list :").split())))
reverse_list=[]
for num in range(len(number_list)-1,-1,-1):
  reverse_list.append(number_list[num])
for i in reverse_list:
  print(i,end=" ")
````
